```python
x = df[["hemoglobin", "smoking"]]
x.head()
```

```python
model= KMeans(n_clusters=3)
y_kmeans = model.fit_predict(x)
```

```python
df['y'] = y_kmeans
```

```python
plt.scatter(df['hemoglobin'], df['smoking'], c=df['y'])
```

```python
model.inertia_
```

```python
wcss = []
for i in range(1,11):
    model= KMeans(n_clusters=i)
    y_kmeans  = model.fit_predict(x)
    wcss.append(model.inertia_)
```

```python
plt.plot(range(1,11), wcss)
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.title('Elbow Method')
plt.show()
```
